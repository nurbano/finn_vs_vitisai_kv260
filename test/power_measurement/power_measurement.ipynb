```python
import matplotlib.pyplot as plt
from time import time
import pandas as pd
```

```python
path="/sys/class/hwmon/hwmon1/"
file="power1_input"
path+file
```

```python
f = open(path+file, "r")
print(f.read()) 
```

```python
power=[]
t=[]
init= time()
ventana= 1
t_total= 30
actual= time()
while(ventana):

    if time()>actual+0.1:
        t.append(time()-init)
        f = open(path+file, "r")
        po= int(f.read())/1000
        power.append(po)
#         print(po,round(actual-init,2))
        actual= time()
        
    dt= time()-init
    if dt>t_total:
        ventana=0
```

```python
plt.plot(t, power)
```

```python
df=pd.DataFrame(data={"tiempo": t, "potencia": power})
df.to_csv()
```

```python
import numpy as np
 
arr = power
window_size = 4
 
i = 0
# Initialize an empty list to store moving averages
moving_averages = []
 
# Loop through the array t o
#consider every window of size 3
while i < len(arr) - window_size + 1:
 
    # Calculate the average of current window
    window_average = round(np.sum(arr[
      i:i+window_size]) / window_size, 2)
     
    # Store the average of current
    # window in moving average list
    moving_averages.append(window_average)
     
    # Shift window to right by one position
    i += 1
```

```python
plt.plot(moving_averages)
```
